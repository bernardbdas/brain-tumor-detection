```python
import sys
import os
import platform
import glob

VENV_DIR = "../.venv" 

def inject_venv_path(venv_name):
    site_packages = None
    
    if platform.system() == "Windows":
        site_packages = os.path.join(venv_name, "lib", "site-packages")
    else:
        # Linux/macOS path structure: venv/lib/python3.x/site-packages
        lib_dirs = glob.glob(os.path.join(venv_name, "lib", "python3.*", "site-packages"))
        if lib_dirs:
            site_packages = lib_dirs[0]

    if site_packages and os.path.exists(site_packages):
        abs_path = os.path.abspath(site_packages)
        if abs_path not in sys.path:
            # Insert at the beginning of sys.path to prioritize these packages
            sys.path.insert(0, abs_path)
        print(f"✅ Successfully injected venv path: {abs_path}")
    else:
        print(f"⚠️ Could not find site-packages in '{venv_name}'.")
        print("Ensure the folder exists in the same directory as this notebook.")

inject_venv_path(VENV_DIR)
```

```python
print("--- PyTorch CUDA Check ---")
try:
    import torch
    print(f"PyTorch Version: {torch.__version__}")
    
    cuda_available = torch.cuda.is_available()
    print(f"CUDA Available: {cuda_available}")
    
    if cuda_available:
        print(f"CUDA Version (built with): {torch.version.cuda}")
        print(f"Device Count: {torch.cuda.device_count()}")
        print(f"Current Device Name: {torch.cuda.get_device_name(torch.cuda.current_device())}")
except ImportError:
    print("❌ PyTorch is not installed in the detected environment.")
```

```python
print("--- TensorFlow CUDA Check ---")
try:
    # Suppress excessive TF logging for a cleaner output
    os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' 
    import tensorflow as tf
    print(f"TensorFlow Version: {tf.__version__}")
    
    built_with_cuda = tf.test.is_built_with_cuda()
    print(f"Built with CUDA: {built_with_cuda}")
    
    gpu_devices = tf.config.list_physical_devices('GPU')
    if gpu_devices:
        print(f"GPU Devices Found: {len(gpu_devices)}")
        for i, device in enumerate(gpu_devices):
            print(f"  Device {i}: {device.name}")
    else:
        print("GPU Devices Found: 0")
except ImportError:
    print("❌ TensorFlow is not installed in the detected environment.")
```

```python
print("--- JAX CUDA Check ---")
try:
    import jax
    print(f"JAX Version: {jax.__version__}")
    
    # JAX defaults to CPU if it can't find a GPU/TPU
    backend = jax.default_backend()
    print(f"Default Backend: {backend.upper()}")
    
    devices = jax.devices()
    print(f"Devices Found: {len(devices)}")
    for i, device in enumerate(devices):
        print(f"  Device {i}: {device.device_kind} ({device})")
        
    if backend != "gpu":
         print("⚠️ JAX is using CPU. Ensure 'jaxlib' is installed with CUDA support.")
except ImportError:
    print("❌ JAX is not installed in the detected environment.")
```
